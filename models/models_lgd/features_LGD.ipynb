```python
import pandas as pd

pd.set_option('display.max_columns', 500)

df = pd.read_csv(
    "/home/muzzi/enterprise_credit_risk_forecasting/data/accepted_2007_to_2018q4.csv/accepted_2007_to_2018Q4.csv",
    low_memory=False,
    nrows=50000
)

print(df.shape)

df.head()
```

```python
# ============================================================
# ENTERPRISE LGD FEATURE ENGINEERING PIPELINE
# ============================================================
# GOAL:
# Build robust LGD modeling dataset
# WITHOUT leakage.
#
# LGD = Loss Given Default
#
# Predict:
# "How much money is lost IF borrower defaults?"
#
# IMPORTANT:
# ONLY use origination-time information.
#
# DO NOT USE:
# - recoveries
# - collections
# - settlements
# - hardship outcomes
# - future balances
# - future payments
#
# ============================================================

import numpy as np
import pandas as pd

# ============================================================
# COPY DATA
# ============================================================

lgd_df = df.copy()

# ============================================================
# KEEP ONLY DEFAULTED LOANS
# ============================================================
# LGD only applies to bad/defaulted loans

bad_status = [

    "Charged Off",

    "Default"
]

lgd_df = lgd_df[

    lgd_df["loan_status"].isin(bad_status)

].copy()

# ============================================================
# CREATE LGD TARGET
# ============================================================

lgd_df["LGD"] = (

    1

    - (

        lgd_df["total_rec_prncp"]

        / (

            lgd_df["funded_amnt"]

            + 1
        )
    )
)

# ============================================================
# CLIP TARGET
# ============================================================

lgd_df["LGD"] = (

    lgd_df["LGD"]

    .clip(0, 1)
)

# ============================================================
# DATE FEATURES
# ============================================================

lgd_df["issue_d"] = pd.to_datetime(

    lgd_df["issue_d"],

    format="%b-%Y",

    errors="coerce"
)

lgd_df["earliest_cr_line"] = pd.to_datetime(

    lgd_df["earliest_cr_line"],

    format="%b-%Y",

    errors="coerce"
)

# ============================================================
# CREDIT HISTORY
# ============================================================

lgd_df["credit_history_months"] = (

    (

        lgd_df["issue_d"]

        - lgd_df["earliest_cr_line"]
    )

    .dt.days

    / 30
)

# ============================================================
# TERM FEATURE
# ============================================================

lgd_df["term_months"] = (

    lgd_df["term"]

    .str.extract(r"(\d+)")

    .astype(float)
)

# ============================================================
# EMPLOYMENT LENGTH
# ============================================================

lgd_df["emp_length_years"] = (

    lgd_df["emp_length"]

    .str.extract(r"(\d+)")

    .astype(float)
)

lgd_df["emp_length_years"] = (

    lgd_df["emp_length_years"]

    .fillna(0)
)

# ============================================================
# BASIC AFFORDABILITY FEATURES
# ============================================================

# Monthly income
lgd_df["monthly_income"] = (

    lgd_df["annual_inc"] / 12
)

# Loan burden
lgd_df["loan_to_income"] = (

    lgd_df["loan_amnt"]

    / (

        lgd_df["annual_inc"]

        + 1
    )
)

# Installment burden
lgd_df["installment_to_income"] = (

    lgd_df["installment"]

    / (

        lgd_df["monthly_income"]

        + 1
    )
)

# Revolving burden
lgd_df["revolving_to_income"] = (

    lgd_df["revol_bal"]

    / (

        lgd_df["annual_inc"]

        + 1
    )
)

# Debt saturation
lgd_df["debt_saturation"] = (

    lgd_df["total_bal_ex_mort"]

    / (

        lgd_df["annual_inc"]

        + 1
    )
)

# ============================================================
# UTILIZATION / LIQUIDITY FEATURES
# ============================================================

# Liquidity exhaustion
lgd_df["liquidity_exhaustion"] = (

    lgd_df["revol_bal"]

    / (

        lgd_df["total_rev_hi_lim"]

        + 1
    )
)

# Revolving utilization stress
lgd_df["utilization_dti_stress"] = (

    lgd_df["revol_util"]

    * lgd_df["dti"]
)

# Bankcard stress
lgd_df["bcutil_dti_stress"] = (

    lgd_df["bc_util"]

    * lgd_df["dti"]
)

# Utilization per account
lgd_df["util_per_account"] = (

    lgd_df["revol_bal"]

    / (

        lgd_df["open_acc"]

        + 1
    )
)

# Remaining revolving capacity
lgd_df["remaining_revolving_capacity"] = (

    lgd_df["total_rev_hi_lim"]

    - lgd_df["revol_bal"]
)

# Remaining bankcard capacity
lgd_df["remaining_bc_capacity"] = (

    lgd_df["total_bc_limit"]

    - (

        (

            lgd_df["bc_util"]

            / 100
        )

        * lgd_df["total_bc_limit"]
    )
)

# ============================================================
# CREDIT HUNGER FEATURES
# ============================================================

# Credit hunger
lgd_df["credit_hunger"] = (

    lgd_df["inq_last_6mths"]

    + lgd_df["acc_open_past_24mths"]
)

# Inquiry intensity
lgd_df["inq_per_year"] = (

    lgd_df["inq_last_6mths"]

    / (

        (

            lgd_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

# Recent account pressure
lgd_df["recent_account_pressure"] = (

    lgd_df["num_tl_op_past_12m"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Leverage acceleration
lgd_df["leverage_acceleration"] = (

    lgd_df["revol_util"]

    * lgd_df["acc_open_past_24mths"]

    * (

        lgd_df["inq_last_6mths"]

        + 1
    )
)

# Utilization shock risk
lgd_df["utilization_shock_risk"] = (

    lgd_df["revol_util"]

    * (

        lgd_df["num_tl_op_past_12m"]

        + 1
    )
)

# ============================================================
# CREDIT STRUCTURE FEATURES
# ============================================================

# Average account age
lgd_df["avg_account_age"] = (

    lgd_df["credit_history_months"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Trade density
lgd_df["trade_density"] = (

    lgd_df["total_acc"]

    / (

        (

            lgd_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

# Credit velocity
lgd_df["credit_velocity"] = (

    lgd_df["total_acc"]

    / (

        (

            lgd_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

# Credit fatigue
lgd_df["credit_fatigue"] = (

    lgd_df["num_actv_rev_tl"]

    / (

        (

            lgd_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

# Revolving dependency
lgd_df["revolving_dependency"] = (

    lgd_df["revol_bal"]

    / (

        lgd_df["total_bal_ex_mort"]

        + 1
    )
)

# Revolving trade share
lgd_df["revolving_trade_share"] = (

    lgd_df["num_rev_accts"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Installment trade share
lgd_df["installment_trade_share"] = (

    lgd_df["num_il_tl"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Active trade ratio
lgd_df["active_trade_ratio"] = (

    lgd_df["num_actv_rev_tl"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Open account ratio
lgd_df["open_account_ratio"] = (

    lgd_df["open_acc"]

    / (

        lgd_df["total_acc"]

        + 1
    )
)

# Account fragmentation
lgd_df["account_fragmentation"] = (

    lgd_df["open_acc"]

    / (

        lgd_df["num_rev_accts"]

        + 1
    )
)

# Borrower complexity
lgd_df["borrower_complexity"] = (

    lgd_df["open_acc"]

    + lgd_df["num_rev_accts"]

    + lgd_df["num_il_tl"]

    + lgd_df["acc_open_past_24mths"]
)

# ============================================================
# FICO INTERACTION FEATURES
# ============================================================

# DTI adjusted by FICO
lgd_df["dti_per_fico"] = (

    lgd_df["dti"]

    / (

        lgd_df["fico_range_low"]

        + 1
    )
)

# Loan burden adjusted by FICO
lgd_df["loan_per_fico"] = (

    lgd_df["loan_amnt"]

    / (

        lgd_df["fico_range_low"]

        + 1
    )
)

# Utilization adjusted by FICO
lgd_df["utilization_per_fico"] = (

    lgd_df["revol_util"]

    / (

        lgd_df["fico_range_low"]

        + 1
    )
)

# FICO stress
lgd_df["fico_stress"] = (

    lgd_df["revol_util"]

    * lgd_df["dti"]

    / (

        lgd_df["fico_range_low"]

        + 1
    )
)

# Credit resilience
lgd_df["credit_resilience"] = (

    lgd_df["credit_history_months"]

    * (

        lgd_df["fico_range_low"]

        + 1
    )

    / (

        lgd_df["dti"]

        + 1
    )
)

# ============================================================
# DELINQUENCY FEATURES
# ============================================================

# Delinquency intensity
lgd_df["delinq_intensity"] = (

    lgd_df["delinq_2yrs"]

    / (

        (

            lgd_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

# Derogatory pressure
lgd_df["derogatory_pressure"] = (

    lgd_df["pub_rec"]

    + lgd_df["pub_rec_bankruptcies"]
)

# ============================================================
# TEMPORAL / LONG-TERM STRESS
# ============================================================

# Long-term affordability stress
lgd_df["long_term_stress"] = (

    lgd_df["term_months"]

    * lgd_df["installment_to_income"]
)

# Long-term leverage burden
lgd_df["term_loan_burden"] = (

    lgd_df["term_months"]

    * lgd_df["loan_to_income"]
)

# Payment stress amplification
lgd_df["payment_stress_amplification"] = (

    lgd_df["installment_to_income"]

    * lgd_df["dti"]
)

# ============================================================
# ISSUE DATE TEMPORAL FEATURES
# ============================================================

lgd_df["issue_year"] = (

    lgd_df["issue_d"]

    .dt.year
)

lgd_df["issue_month"] = (

    lgd_df["issue_d"]

    .dt.month
)

lgd_df["issue_quarter"] = (

    lgd_df["issue_d"]

    .dt.quarter
)

# ============================================================
# REMOVE MASSIVE LEAKAGE COLUMNS
# ============================================================

lgd_leakage_cols = [

    # direct recovery leakage
    "recoveries",
    "collection_recovery_fee",
    "total_rec_prncp",
    "total_rec_int",
    "total_rec_late_fee",
    "total_pymnt",
    "total_pymnt_inv",

    # future balances
    "out_prncp",
    "out_prncp_inv",

    # settlement leakage
    "settlement_status",
    "settlement_date",
    "settlement_amount",
    "settlement_percentage",
    "settlement_term",
    "debt_settlement_flag",
    "debt_settlement_flag_date",

    # hardship leakage
    "hardship_flag",
    "hardship_type",
    "hardship_reason",
    "hardship_status",
    "hardship_amount",
    "hardship_start_date",
    "hardship_end_date",
    "payment_plan_start_date",
    "hardship_length",
    "hardship_dpd",
    "hardship_loan_status",
    "orig_projected_additional_accrued_interest",
    "hardship_payoff_balance_amount",
    "hardship_last_payment_amount",

    # future payment information
    "last_pymnt_d",
    "last_pymnt_amnt",
    "next_pymnt_d",
    "last_credit_pull_d",
    "last_fico_range_high",
    "last_fico_range_low",

    # direct identifiers
    "id",
    "member_id",
    "url",
    "desc",

    # target leakage
    "loan_status"
]

lgd_df.drop(

    columns=lgd_leakage_cols,

    errors="ignore",

    inplace=True
)

# ============================================================
# CLEAN INFINITE VALUES
# ============================================================

lgd_df = lgd_df.replace(

    [np.inf, -np.inf],

    np.nan
)

# ============================================================
# FINAL INFO
# ============================================================

print("✅ LGD Feature Engineering Complete")

print(f"Rows    : {lgd_df.shape[0]}")

print(f"Columns : {lgd_df.shape[1]}")

print(f"Mean LGD: {lgd_df['LGD'].mean():.4f}")

print(f"Median LGD: {lgd_df['LGD'].median():.4f}")
```

### 📊 LGD Feature Dictionary & Mathematical Specifications

The table below summarizes a highly predictive selection of the **151 features** utilized in the state-of-the-art CatBoost Loss Given Default (LGD) model. Because this LGD regression pipeline operates exclusively on accounts that have already failed (`TARGET == 1`), these features are structured to isolate non-linear cash-flow strains, structural leverage cliffs, and borrower financial complexity.


| Feature Name | Core Business Meaning | Mathematical Formula / Algorithmic Representation |
| :--- | :--- | :--- |
| **`loan_amnt`** | The total nominal value of the loan requested by the applicant. | Input Column: $A_{loan} \in \mathbb{R}^{+}$ |
| **`installment`** | The fixed contractual monthly payment owed by the borrower. | $I = A_{loan} \times \frac{r(1+r)^N}{(1+r)^N - 1}$ <br> *(Where $r = \text{monthly interest rate}$, $N = \text{term months}$)* |
| **`fico_average`** | The synthesized credit score baseline representing creditworthiness. | $\text{FICO}_{avg} = \frac{\text{fico\_range\_low} + \text{fico\_range\_high}}{2}$ |
| **`dti`** | Debt-to-Income ratio; evaluates core borrower debt concentration. | $\text{DTI} = \frac{\text{Total Monthly Debt Obligations (Excl. Mortgage)}}{\text{Gross Monthly Income}}$ |
| **`credit_history_months`** | Total depth and maturity of the borrower's credit footprint. | $T_{history} = \text{issue\_d} - \text{earliest\_cr\_line}$ (expressed in calendar months) |
| **`term_months`** | Total duration of the credit contract (e.g., 36 or 60 months). | Input Column: $M_{term} \in \{36, 60\}$ |
| **`monthly_income`** | The normalized monthly earning potential of the applicant. | $I_{monthly} = \frac{\text{annual\_inc}}{12}$ |
| **`loan_to_income`** | Leverage ratio; sizes the total principal obligation against annual salary. | $\text{LTI} = \frac{\text{loan\_amnt}}{\text{annual\_inc}}$ |
| **`installment_to_income`** | The immediate contractual cash drain on the borrower's monthly paycheck. | $\text{ITI} = \frac{\text{installment}}{I_{monthly}}$ |
| **`revolving_to_income`** | Measures how much active credit card debt balances swallow annual income. | $\text{RTI} = \frac{\text{revol\_bal}}{\text{annual\_inc}}$ |
| **`debt_saturation`** | Structural leverage stress tracking total credit usage relative to capacity. | $\text{DS} = \text{dti} \times \text{revol\_util}$ |
| **`liquidity_exhaustion`** | Calculates how fast an applicant will run completely out of cash reserves. | $\text{LE} = \frac{\text{installment} + \text{revol\_bal}}{\text{annual\_inc}}$ |
| **`utilization_dti_stress`** | Scales debt-to-income limits by current line utilization constraints. | $\text{UDTS} = \text{dti} \times (1 + \text{revol\_util})$ |
| **`bcutil_dti_stress`** | Measures debt-to-income limits adjusted exclusively by bankcard strain. | $\text{BDTS} = \text{dti} \times (1 + \text{bc\_util})$ |
| **`util_per_account`** | Identifies if balances are evenly spread or dangerously maxed on single lines. | $\text{UPA} = \frac{\text{revol\_util}}{\text{open\_acc} + 1}$ |
| **`remaining_revolving_capacity`**| Total liquid cash cushion accessible via active revolving credit networks.| $\text{RRC} = \text{total\_rev\_hi\_lim} - \text{revol\_bal}$ |
| **`remaining_bc_capacity`** | Total unutilized credit safety net left specifically on bank credit cards. | $\text{BCC} = \text{total\_bc\_limit} - (\text{total\_bc\_limit} \times \text{bc\_util})$ |
| **`credit_hunger`** | Captures urgent credit-seeking velocity over the last 6 months. | $\text{CH} = \frac{\text{inq\_last\_6mths}}{\text{open\_acc} + 1}$ |
| **`inq_per_year`** | Annualized inquiry rate; spikes signify worsening risk profiles. | $\text{IPY} = \frac{\text{inq\_last\_12m}}{\max(1, T_{history}/12)}$ |
| **`recent_account_pressure`** | Evaluates immediate payment strain from newly opened credit accounts. | $\text{RAP} = \frac{\text{acc\_open\_past\_24mths}}{\text{total\_acc} + 1}$ |
| **`leverage_acceleration`** | Velocity metric checking if open trade lines are climbing unsustainably. | $\text{LA} = \frac{\text{num\_op\_rev\_tl}}{\text{num\_rev\_accts} + 1}$ |
| **`utilization_shock_risk`** | Flags fragile profiles highly exposed to sudden interest or utilization shifts. | $\text{USR} = \frac{\text{all\_util}}{\text{fico\_average}}$ |
| **`avg_account_age`** | The average maturity across all credit lines in the borrower's portfolio. | $\text{AAA} = \frac{\text{mo\_sin\_old\_rev\_tl\_op} + \text{mo\_sin\_old\_il\_acct}}{2}$ |
| **`trade_density`** | Measures structural profile complexity and account fragmentation lines. | $\text{TD} = \frac{\text{open\_acc}}{\text{total\_acc} + 1}$ |
| **`credit_velocity`** | Tracks the rate of newly established trades relative to historic baseline age. | $\text{CV} = \frac{\text{acc\_open\_past\_24mths}}{\text{mo\_sin\_old\_rev\_tl\_op} + 1}$ |
| **`credit_fatigue`** | Proxy for exhausting total borrowing potential over a multi-year window. | $\text{CF} = \frac{\text{total\_acc}}{\text{fico\_average}}$ |
| **`revolving_dependency`** | Tracks the share of active lines locked in unpredictable revolving balances. | $\text{RD} = \frac{\text{num\_actv\_rev\_tl}}{\text{open\_acc} + 1}$ |
| **`revolving_trade_share`** | Measures the percentage of the portfolio tied to credit cards vs. term loans. | $\text{RTS} = \frac{\text{num\_rev\_accts}}{\text{total\_acc} + 1}$ |
| **`installment_trade_share`** | Measures the percentage of the portfolio tied to fixed-term installment loans. | $\text{ITS} = \frac{\text{num\_il\_tl}}{\text{total\_acc} + 1}$ |
| **`active_trade_ratio`** | Tracks how many available revolving lines are actively carrying a balance. | $\text{ATR} = \frac{\text{num\_rev\_tl\_bal\_gt\_0}}{\text{num\_op\_rev\_tl} + 1}$ |
| **`open_account_ratio`** | Measures the proportion of accounts currently active out of total historical accounts. | $\text{OAR} = \frac{\text{open\_acc}}{\text{total\_acc} + 1}$ |
| **`account_fragmentation`** | Tracks borrower reliance on multiple simultaneous credit suppliers. | $\text{AF} = \frac{\text{open\_acc}}{\text{num\_sats} + 1}$ |
| **`borrower_complexity`** | Captures financial entanglement via the count of overall active trade profiles. | $\text{BC} = \text{open\_acc} + \text{num\_rev\_accts} + \text{num\_il\_tl}$ |
| **`dti_per_fico`** | Risk-adjusted debt layer sizing traditional DTI burden against credit tier. | $\text{DTIF} = \frac{\text{dti}}{\text{fico\_average}}$ |
| **`loan_per_fico`** | Sizes requested asset concentration relative to historical credit tiering. | $\text{LPF} = \frac{\text{loan\_amnt}}{\text{fico\_average}}$ |
| **`utilization_per_fico`** | Normalizes current utilization strain relative to the borrower's FICO rank. | $\text{UPF} = \frac{\text{all\_util}}{\text{fico\_average}}$ |
| **`fico_stress`** | Mathematical penalty sizing credit degradation underneath baseline limits. | $\text{FS} = \frac{850 - \text{fico\_average}}{850 - 300}$ |
| **`credit_resilience`** | Measures clean payment history longevity against continuous credit use. | $\text{CR} = \frac{\text{pct\_tl\_nvr\_dlq}}{\text{all\_util} + 1}$ |
| **`delinq_intensity`** | Tracks historical delinquency severity against active open channels. | $\text{DI} = \frac{\text{delinq\_2yrs}}{\text{open\_acc} + 1}$ |
| **`delinq\_intensity`** | Tracks historical delinquency severity against active open channels. | $\text{DI} = \frac{\text{delinq\_2yrs}}{\text{open\_acc} + 1}$ |
| **`derogatory_pressure`** | Captures severe legal or public credit wreckage relative to account depth. | $\text{DP} = \frac{\text{pub\_rec} + \text{pub\_rec\_bankruptcies}}{\text{total\_acc} + 1}$ |

```python
import numpy as np
import pandas as pd
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# ==========================================
# 1. SETUP DATA AND CONFIGURATIONS
# ==========================================
# Assuming `lgd_dataset` is your pre-filtered DataFrame where TARGET == 1

# Features that must be excluded to prevent data loops and syntax crashes
unwanted_cols = [
    'TARGET', 'LGD',                      # Direct machine learning targets
    'issue_d', 'earliest_cr_line',        # Raw unparsed datetime strings
    'sec_app_earliest_cr_line'             # Secondary unparsed datetime string
]

# Ensure LGD contains no missing values for the regression target
lgd_dataset = lgd_dataset.dropna(subset=['LGD']).copy()

# Dynamically extract all valid predictors based strictly on your printed list
features_to_use = [col for col in lgd_dataset.columns if col not in unwanted_cols]
X = lgd_dataset[features_to_use]
y = lgd_dataset['LGD'].astype(float)

# Discover categoricals and force clean string casting to prevent TypeErrors
cat_cols = X.select_dtypes(include=['object', 'category', 'string']).columns.tolist()
for col in cat_cols:
    X[col] = X[col].astype(str).fillna('missing')

print(f"✅ Data processed successfully for LGD Analysis.")
print(f"📊 Matrix Breakdown: {X.shape[0]} Defaulted Loans | {X.shape[1]} Predictors.")
print(f"🗂️ Categorical Feature Total: {len(cat_cols)}")

# ==========================================
# 2. DEFINING SOTA LGD HYPERPARAMETERS
# ==========================================
# Bounded continuous spaces require high L2 regularisation to steady the leaf targets
sota_lgd_params = {
    'iterations': 1200,                  # Budgets early stopping checks
    'learning_rate': 0.02,               # Tight optimization step
    'depth': 5,                          # Shallow structures protect narrow defaulted row sample sets
    'l2_leaf_reg': 25,                   # Heavy penalty mitigates feature dilution over 151 variables
    'loss_function': 'RMSE',             # Root Mean Squared Error optimizes baseline variance
    'eval_metric': 'R2',                 # Tracks the actual coefficient of determination
    'random_seed': 42,
    'early_stopping_rounds': 50,
    'verbose': 0                         # Clean outputs across cross-validation blocks
}

# ==========================================
# 3. K-FOLD CROSS-VALIDATION LOOP
# ==========================================
# Shuffled K-Fold replaces Stratified splits because the target variable is continuous
kf = KFold(n_splits=5, shuffle=True, random_state=42)

fold_metrics = []
oof_predictions = np.zeros(len(X))

print("\n🚀 Commencing 5-Fold LGD Regression Run...")

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y), 1):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Pack parameters inside native arrays
    train_pool = Pool(X_train, y_train, cat_features=cat_cols)
    val_pool = Pool(X_val, y_val, cat_features=cat_cols)
    
    # Note: Using CatBoostRegressor here instead of Classifier
    model = CatBoostRegressor(**sota_lgd_params)
    model.fit(train_pool, eval_set=val_pool, use_best_model=True)
    
    # Predict and strictly clip to logical financial bounds [0%, 100%]
    val_preds = np.clip(model.predict(val_pool), 0.0, 1.0)
    oof_predictions[val_idx] = val_preds
    
    # Calculate performance indices
    r2 = r2_score(y_val, val_preds)
    rmse = np.sqrt(mean_squared_error(y_val, val_preds))
    mae = mean_absolute_error(y_val, val_preds)
    
    fold_metrics.append({'R2': r2, 'RMSE': rmse, 'MAE': mae})
    print(f"  ↳ Fold {fold} Metrics -> R²: {r2:.4f} | RMSE: {rmse:.4f} | MAE: {mae:.4f}")

# ==========================================
# 4. FINAL REGRESSION PERFORMANCE SUMMARY
# ==========================================
print("\n================================================================================")
print("FINAL LGD MODEL SUMMARY")
print("================================================================================")
print(f"MEAN R² SCORE (Explained Variance) : {np.mean([f['R2'] for f in fold_metrics]):.4f}")
print(f"MEAN RMSE (Avg Deviation Error)    : {np.mean([f['RMSE'] for f in fold_metrics]):.4f}")
print(f"MEAN MAE (Mean Absolute Error)     : {np.mean([f['MAE'] for f in fold_metrics]):.4f}")
print(f"R² ST. DEV.                        : {np.std([f['R2'] for f in fold_metrics]):.4f}")
print("================================================================================")
```
