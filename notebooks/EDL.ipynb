```python
## all_util — 57% missing ⭐⭐⭐⭐⭐ Combined utilization metric. IMPORTANT QUESTION ⭐⭐⭐⭐⭐ We must later investigate: Is missingness random? OR correlated with:

# HYPOTHESIS TESTS ⭐⭐⭐⭐⭐

# We want to test:

# Hypothesis	Meaning
# missing random	no business meaning
# missing = thin file	customer lacks enough credit history
# missing = low utilization	no active debt
# missing = risky customers	hidden deterioration
# missing = safe customers	financially inactive


# ============================================================
# CLEAN TARGET CREATION
# ============================================================

bad_status = [

    "Charged Off",

    "Default"
]

good_status = [

    "Fully Paid"
]

df_pd = df[

    df["loan_status"].isin(
        bad_status + good_status
    )

].copy()

df_pd["TARGET"] = (

    df_pd["loan_status"]
    .isin(bad_status)

).astype(int)

# ============================================================
# TARGET DISTRIBUTION
# ============================================================

print(

    df_pd["TARGET"]
    .value_counts(normalize=True)

)
```

```python
# STEP 1 — CREATE MISSING FLAG ⭐⭐⭐⭐⭐
# ============================================================
# MISSING FLAG
# ============================================================

## target label

df_pd["ALL_UTIL_MISSING"] = (

    df_pd["all_util"]
    .isnull()
).astype(int)

# CHECK DISTRIBUTION
df_pd["ALL_UTIL_MISSING"].value_counts()
```

```python
# STEP 2 — CHECK DEFAULT RATE ⭐⭐⭐⭐⭐

# MOST IMPORTANT TEST.

# ============================================================
# DEFAULT RATE BY MISSINGNESS
# ============================================================

(
    df_pd
    .groupby("ALL_UTIL_MISSING")["TARGET"]
    .mean()
    * 100
)
# INTERPRETATION ⭐⭐⭐⭐⭐

# Suppose:

# Missing?	Default Rate
# 0	14%
# 1	31%

# strongly predicts risk.
# ALTERNATIVE POSSIBILITY ⭐⭐⭐⭐⭐
# Missing?	Default Rate
# 0	18%
# 1	5%

# Then:
# missing may imply:

# low leverage,
# financially inactive customers.

# VERY important.
```

```python
# STEP 3 — CHECK THIN FILE HYPOTHESIS ⭐⭐⭐⭐⭐

# We suspect:
# missing may correlate with:

# low credit history.

# Test:

# ============================================================
# CREDIT HISTORY DEPTH
# ============================================================

(
    df_pd
    .groupby("ALL_UTIL_MISSING")[

        [
            "total_acc",

            "open_acc",

            "mort_acc"
        ]
    ]
    .mean()
)
# INTERPRETATION ⭐⭐⭐⭐⭐

# If missing customers have:

# Feature	Lower?
# total_acc	YES
# open_acc	YES

# Then:
# missing likely means:

# thin-file borrowers.
```

```python
# 1. STRUCTURAL MISSING ⭐⭐⭐⭐⭐

# Meaning:

# feature not applicable

# Usually:

# joint application absent,
# installment subsystem absent,
# product not used.
# FEATURES ⭐⭐⭐⭐⭐
STRUCTURAL_MISSING = [

    # JOINT APPLICATION
    "annual_inc_joint",
    "dti_joint",
    "verification_status_joint",

    # SECONDARY APPLICANT
    "sec_app_fico_range_low",
    "sec_app_fico_range_high",
    "sec_app_earliest_cr_line",
    "sec_app_inq_last_6mths",
    "sec_app_mort_acc",
    "sec_app_open_acc",
    "sec_app_revol_util",
    "sec_app_open_act_il",
    "sec_app_num_rev_accts",
    "sec_app_chargeoff_within_12_mths",
    "sec_app_collections_12_mths_ex_med",
    "sec_app_mths_since_last_major_derog",

    # INSTALLMENT SUBSYSTEM
    "il_util",
    "mths_since_rcnt_il",
    "open_il_12m",
    "open_il_24m",
    "open_act_il",

]
```

```python
# STRATEGY ⭐⭐⭐⭐⭐
# ============================================================
# STRUCTURAL MISSING
# ============================================================

for col in STRUCTURAL_MISSING:

    # CREATE INDICATOR
    df_pd[f"{col}_MISSING"] = (

        df_pd[col]
        .isnull()
    ).astype(int)

    # SAFE IMPUTATION
    df_pd[col] = df_pd[col].fillna(-999)
# WHY -999? ⭐⭐⭐⭐⭐

# For tree models:

# creates separable branch,
# preserves missing signal,
# avoids false meaning.
```

```python
# 2. BEHAVIORAL MISSING ⭐⭐⭐⭐⭐

# Meaning:

# missing itself carries risk meaning

# VERY important.

# FEATURES ⭐⭐⭐⭐⭐
BEHAVIORAL_MISSING = [

    "mths_since_last_delinq",
    "mths_since_last_record",
    "mths_since_last_major_derog",
    "mths_since_recent_bc_dlq",
    "mths_since_recent_revol_delinq",

    # HARDSHIP
    "hardship_reason",
    "hardship_type",
    "hardship_status",

    # SETTLEMENT
    "settlement_status",
    "settlement_amount",
    "settlement_percentage",
]
```

```python
# STRATEGY ⭐⭐⭐⭐⭐
# ============================================================
# BEHAVIORAL MISSING
# ============================================================

for col in BEHAVIORAL_MISSING:

    # MISSING FLAG
    df_pd[f"{col}_MISSING"] = (

        df_pd[col]
        .isnull()
    ).astype(int)

# SPECIAL TIME-SINCE HANDLING
TIME_SINCE_COLS = [

    "mths_since_last_delinq",
    "mths_since_last_record",
    "mths_since_last_major_derog",
    "mths_since_recent_bc_dlq",
    "mths_since_recent_revol_delinq",
]

for col in TIME_SINCE_COLS:

    # NEVER EVENT
    df_pd[col] = df_pd[col].fillna(999)
# WHY 999? ⭐⭐⭐⭐⭐

# Because:

# missing often means:
# event never happened

# 999 approximates:

# very long ago / effectively never

# Much safer than:

# fillna(0)
```

```python
# 3. LEAKAGE FEATURES ⭐⭐⭐⭐⭐

# Meaning:

# future information unavailable at decision time

# VERY dangerous.

# FEATURES ⭐⭐⭐⭐⭐
LEAKAGE_FEATURES = [

    "loan_status",

    "out_prncp",
    "out_prncp_inv",

    "total_pymnt",
    "total_pymnt_inv",

    "total_rec_prncp",
    "total_rec_int",
    "total_rec_late_fee",

    "recoveries",
    "collection_recovery_fee",

    "last_pymnt_d",
    "last_pymnt_amnt",

    "next_pymnt_d",

    "debt_settlement_flag",
    "debt_settlement_flag_date",

    "settlement_status",
    "settlement_date",
    "settlement_amount",
    "settlement_percentage",
    "settlement_term",

    "hardship_status",
    "hardship_start_date",
    "hardship_end_date",

]
```

```python
# STRATEGY ⭐⭐⭐⭐⭐
# ============================================================
# REMOVE LEAKAGE
# ============================================================

df_pd = df_pd.drop(

    columns=LEAKAGE_FEATURES,

    errors="ignore"
)
# VERY IMPORTANT ⭐⭐⭐⭐⭐

# For:

# LGD,
# post-default recovery,
# collections modeling,

# some of these become VALID.

# But for:

# PD

# they are illegal.
```

```python
# 4. RANDOM / DATA QUALITY MISSING ⭐⭐⭐⭐⭐

# Meaning:

# missing likely accidental / operational
# FEATURES ⭐⭐⭐⭐⭐
RANDOM_MISSING = [

    "all_util",
    "avg_cur_bal",
    "bc_open_to_buy",
    "bc_util",
    "max_bal_bc",
]
```

```python
# STRATEGY ⭐⭐⭐⭐⭐
# ============================================================
# RANDOM / OPERATIONAL MISSING
# ============================================================

for col in RANDOM_MISSING:

    median_value = df_pd[col].median()

    df_pd[col] = df_pd[col].fillna(
        median_value
    )
```

```python
# 5. OPERATIONAL MISSING ⭐⭐⭐⭐⭐

# Meaning:

# subsystem unavailable
# or reporting inconsistency
# FEATURES ⭐⭐⭐⭐⭐
OPERATIONAL_MISSING = [

    "all_util",
    "total_cu_tl",
    "inq_fi",
]
```

```python
# STRATEGY ⭐⭐⭐⭐⭐
# ============================================================
# OPERATIONAL MISSING
# ============================================================

for col in OPERATIONAL_MISSING:

    df_pd[f"{col}_MISSING"] = (

        df_pd[col]
        .isnull()
    ).astype(int)

    df_pd[col] = df_pd[col].fillna(
        df_pd[col].median()
    )
# ENTERPRISE GOVERNANCE INSIGHT ⭐⭐⭐⭐⭐

# You now have:

# Governance Layer	Purpose
# structural	product applicability
# behavioral	risk signal
# leakage	regulatory safety
# operational	system gaps
# random	statistical imputation

# This is EXACTLY how mature risk organizations think.
```
