```python
import pandas as pd

pd.set_option('display.max_columns', 500)

df = pd.read_csv(
    "/home/muzzi/enterprise_credit_risk_forecasting/data/accepted_2007_to_2018q4.csv/accepted_2007_to_2018Q4.csv",
    low_memory=False,
    nrows=50000
)

print(df.shape)

df.head()
```

```python
# ============================================================
# ENTERPRISE EAD FEATURE ENGINEERING PIPELINE
# ============================================================
# GOAL:
# Build Exposure At Default (EAD) modeling dataset
# WITHOUT leakage.
#
# EAD = Exposure At Default
#
# Predict:
# "How much exposure remains when borrower defaults?"
#
# TARGET:
#
# EAD_ratio = out_prncp / funded_amnt
#
# IMPORTANT:
# - ONLY origination-time information allowed
# - NO future recoveries
# - NO collections
# - NO settlements
# - NO future payment behavior
#
# ============================================================

import numpy as np
import pandas as pd

# ============================================================
# COPY DATA
# ============================================================

ead_df = df.copy()

# ============================================================
# KEEP ONLY DEFAULTED LOANS
# ============================================================

bad_status = [

    "Charged Off",

    "Default"
]

ead_df = ead_df[

    ead_df["loan_status"].isin(bad_status)

].copy()

# ============================================================
# CREATE EAD TARGET
# ============================================================

ead_df["EAD_ratio"] = (

    ead_df["out_prncp"]

    / (

        ead_df["funded_amnt"]

        + 1
    )
)

# ============================================================
# CLEAN TARGET
# ============================================================

ead_df["EAD_ratio"] = (

    ead_df["EAD_ratio"]

    .clip(0, 1)
)

# ============================================================
# DATE CONVERSIONS
# ============================================================

ead_df["issue_d"] = pd.to_datetime(

    ead_df["issue_d"],

    format="%b-%Y",

    errors="coerce"
)

ead_df["earliest_cr_line"] = pd.to_datetime(

    ead_df["earliest_cr_line"],

    format="%b-%Y",

    errors="coerce"
)

# ============================================================
# CREDIT HISTORY
# ============================================================

ead_df["credit_history_months"] = (

    (

        ead_df["issue_d"]

        - ead_df["earliest_cr_line"]
    )

    .dt.days

    / 30
)

# ============================================================
# TERM FEATURE
# ============================================================

ead_df["term_months"] = (

    ead_df["term"]

    .str.extract(r"(\d+)")

    .astype(float)
)

# ============================================================
# EMPLOYMENT LENGTH
# ============================================================

ead_df["emp_length_years"] = (

    ead_df["emp_length"]

    .str.extract(r"(\d+)")

    .astype(float)
)

ead_df["emp_length_years"] = (

    ead_df["emp_length_years"]

    .fillna(0)
)

# ============================================================
# BASIC AFFORDABILITY FEATURES
# ============================================================

ead_df["monthly_income"] = (

    ead_df["annual_inc"] / 12
)

ead_df["loan_to_income"] = (

    ead_df["loan_amnt"]

    / (

        ead_df["annual_inc"]

        + 1
    )
)

ead_df["installment_to_income"] = (

    ead_df["installment"]

    / (

        ead_df["monthly_income"]

        + 1
    )
)

ead_df["revolving_to_income"] = (

    ead_df["revol_bal"]

    / (

        ead_df["annual_inc"]

        + 1
    )
)

ead_df["debt_saturation"] = (

    ead_df["total_bal_ex_mort"]

    / (

        ead_df["annual_inc"]

        + 1
    )
)

# ============================================================
# UTILIZATION & LIQUIDITY FEATURES
# ============================================================

ead_df["liquidity_exhaustion"] = (

    ead_df["revol_bal"]

    / (

        ead_df["total_rev_hi_lim"]

        + 1
    )
)

ead_df["utilization_dti_stress"] = (

    ead_df["revol_util"]

    * ead_df["dti"]
)

ead_df["bcutil_dti_stress"] = (

    ead_df["bc_util"]

    * ead_df["dti"]
)

ead_df["util_per_account"] = (

    ead_df["revol_bal"]

    / (

        ead_df["open_acc"]

        + 1
    )
)

ead_df["remaining_revolving_capacity"] = (

    ead_df["total_rev_hi_lim"]

    - ead_df["revol_bal"]
)

ead_df["remaining_bc_capacity"] = (

    ead_df["total_bc_limit"]

    - (

        (

            ead_df["bc_util"]

            / 100
        )

        * ead_df["total_bc_limit"]
    )
)

# ============================================================
# CREDIT HUNGER FEATURES
# ============================================================

ead_df["credit_hunger"] = (

    ead_df["inq_last_6mths"]

    + ead_df["acc_open_past_24mths"]
)

ead_df["inq_per_year"] = (

    ead_df["inq_last_6mths"]

    / (

        (

            ead_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

ead_df["recent_account_pressure"] = (

    ead_df["num_tl_op_past_12m"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["leverage_acceleration"] = (

    ead_df["revol_util"]

    * ead_df["acc_open_past_24mths"]

    * (

        ead_df["inq_last_6mths"]

        + 1
    )
)

ead_df["utilization_shock_risk"] = (

    ead_df["revol_util"]

    * (

        ead_df["num_tl_op_past_12m"]

        + 1
    )
)

# ============================================================
# CREDIT STRUCTURE FEATURES
# ============================================================

ead_df["avg_account_age"] = (

    ead_df["credit_history_months"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["trade_density"] = (

    ead_df["total_acc"]

    / (

        (

            ead_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

ead_df["credit_velocity"] = (

    ead_df["total_acc"]

    / (

        (

            ead_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

ead_df["credit_fatigue"] = (

    ead_df["num_actv_rev_tl"]

    / (

        (

            ead_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

ead_df["revolving_dependency"] = (

    ead_df["revol_bal"]

    / (

        ead_df["total_bal_ex_mort"]

        + 1
    )
)

ead_df["revolving_trade_share"] = (

    ead_df["num_rev_accts"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["installment_trade_share"] = (

    ead_df["num_il_tl"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["active_trade_ratio"] = (

    ead_df["num_actv_rev_tl"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["open_account_ratio"] = (

    ead_df["open_acc"]

    / (

        ead_df["total_acc"]

        + 1
    )
)

ead_df["account_fragmentation"] = (

    ead_df["open_acc"]

    / (

        ead_df["num_rev_accts"]

        + 1
    )
)

ead_df["borrower_complexity"] = (

    ead_df["open_acc"]

    + ead_df["num_rev_accts"]

    + ead_df["num_il_tl"]

    + ead_df["acc_open_past_24mths"]
)

# ============================================================
# FICO INTERACTION FEATURES
# ============================================================

ead_df["dti_per_fico"] = (

    ead_df["dti"]

    / (

        ead_df["fico_range_low"]

        + 1
    )
)

ead_df["loan_per_fico"] = (

    ead_df["loan_amnt"]

    / (

        ead_df["fico_range_low"]

        + 1
    )
)

ead_df["utilization_per_fico"] = (

    ead_df["revol_util"]

    / (

        ead_df["fico_range_low"]

        + 1
    )
)

ead_df["fico_stress"] = (

    ead_df["revol_util"]

    * ead_df["dti"]

    / (

        ead_df["fico_range_low"]

        + 1
    )
)

ead_df["credit_resilience"] = (

    ead_df["credit_history_months"]

    * (

        ead_df["fico_range_low"]

        + 1
    )

    / (

        ead_df["dti"]

        + 1
    )
)

# ============================================================
# DELINQUENCY FEATURES
# ============================================================

ead_df["delinq_intensity"] = (

    ead_df["delinq_2yrs"]

    / (

        (

            ead_df["credit_history_months"]

            / 12
        )

        + 1
    )
)

ead_df["derogatory_pressure"] = (

    ead_df["pub_rec"]

    + ead_df["pub_rec_bankruptcies"]
)

# ============================================================
# TEMPORAL / AMORTIZATION FEATURES
# ============================================================

ead_df["long_term_stress"] = (

    ead_df["term_months"]

    * ead_df["installment_to_income"]
)

ead_df["term_loan_burden"] = (

    ead_df["term_months"]

    * ead_df["loan_to_income"]
)

ead_df["payment_stress_amplification"] = (

    ead_df["installment_to_income"]

    * ead_df["dti"]
)

# ============================================================
# ISSUE DATE TEMPORAL FEATURES
# ============================================================

ead_df["issue_year"] = (

    ead_df["issue_d"]

    .dt.year
)

ead_df["issue_month"] = (

    ead_df["issue_d"]

    .dt.month
)

ead_df["issue_quarter"] = (

    ead_df["issue_d"]

    .dt.quarter
)

# ============================================================
# AMORTIZATION / EXPOSURE FEATURES
# ============================================================

ead_df["installment_per_loan"] = (

    ead_df["installment"]

    / (

        ead_df["loan_amnt"]

        + 1
    )
)

ead_df["term_installment_pressure"] = (

    ead_df["term_months"]

    * ead_df["installment"]
)

ead_df["exposure_pressure"] = (

    ead_df["loan_amnt"]

    * ead_df["revol_util"]
)

ead_df["exposure_income_stress"] = (

    ead_df["loan_amnt"]

    * ead_df["dti"]

    / (

        ead_df["annual_inc"]

        + 1
    )
)

# ============================================================
# REMOVE LEAKAGE FEATURES
# ============================================================

ead_leakage_cols = [

    # TARGETS
    "TARGET",
    

    # DIRECT EAD LEAKAGE
    "out_prncp",
    "out_prncp_inv",

    # RECOVERY LEAKAGE
    "recoveries",
    "collection_recovery_fee",
    "total_rec_prncp",
    "total_rec_int",
    "total_rec_late_fee",

    # PAYMENT LEAKAGE
    "total_pymnt",
    "total_pymnt_inv",
    "last_pymnt_d",
    "last_pymnt_amnt",
    "next_pymnt_d",

    # FUTURE CREDIT INFO
    "last_credit_pull_d",
    "last_fico_range_high",
    "last_fico_range_low",

    # SETTLEMENT LEAKAGE
    "settlement_status",
    "settlement_date",
    "settlement_amount",
    "settlement_percentage",
    "settlement_term",
    "debt_settlement_flag",
    "debt_settlement_flag_date",

    # HARDSHIP LEAKAGE
    "hardship_flag",
    "hardship_type",
    "hardship_reason",
    "hardship_status",
    "hardship_amount",
    "hardship_start_date",
    "hardship_end_date",
    "payment_plan_start_date",
    "hardship_length",
    "hardship_dpd",
    "hardship_loan_status",
    "orig_projected_additional_accrued_interest",
    "hardship_payoff_balance_amount",
    "hardship_last_payment_amount",

    # IDENTIFIERS
    "id",
    "member_id",
    "url",
    "desc",

    # STATUS
    "loan_status"
]

ead_df.drop(

    columns=ead_leakage_cols,

    errors="ignore",

    inplace=True
)

# ============================================================
# CLEAN INFINITE VALUES
# ============================================================

ead_df = ead_df.replace(

    [np.inf, -np.inf],

    np.nan
)

# ============================================================
# FINAL INFO
# ============================================================

print("✅ EAD Feature Engineering Complete")

print(f"Rows    : {ead_df.shape[0]}")
print(f"Columns : {ead_df.shape[1]}")

print(
    f"Mean EAD Ratio: "
    f"{ead_df['EAD_ratio'].mean():.4f}"
)

print(
    f"Median EAD Ratio: "
    f"{ead_df['EAD_ratio'].median():.4f}"
)
```

```python
import numpy as np
import pandas as pd

# Create a clean working copy of your loaded dataset
processed_df = df.copy()

# ==========================================
# 1. GENERATE THE PD TARGET ('TARGET')
# ==========================================
# Inspect the raw 'loan_status' column to create a clean binary tag
# We isolate completed default events (1) vs. completed successful events (0)
status_mapping = {
    'Fully Paid': 0,
    'Charged Off': 1,
    'Default': 1,
    'Does not meet the credit policy. Status:Fully Paid': 0,
    'Does not meet the credit policy. Status:Charged Off': 1
}

processed_df['TARGET'] = processed_df['loan_status'].map(status_mapping)

# Drop any rows where the status doesn't map (e.g., 'Current', 'Late', or missing values)
# These represent ongoing active loans whose final failure risk cannot be determined yet
processed_df = processed_df.dropna(subset=['TARGET']).copy()
processed_df['TARGET'] = processed_df['TARGET'].astype(int)

# ==========================================
# 2. GENERATE THE EAD TARGET ('EAD')
# ==========================================
# EAD is the total funded principal balance minus the principal they successfully paid back
processed_df['EAD'] = processed_df['funded_amnt'] - processed_df['total_rec_prncp']

# Force a lower floor at 0.0 to prevent minor rounding anomalies on fully paid loans
processed_df['EAD'] = processed_df['EAD'].clip(lower=0.0)

# ==========================================
# 3. GENERATE THE LGD TARGET ('LGD')
# ==========================================
# Initialize LGD as NaN globally across all accounts
processed_df['LGD'] = np.nan

# Locate rows where a default event actually occurred (TARGET == 1)
default_mask = (processed_df['TARGET'] == 1) & (processed_df['EAD'] > 0)

# LGD Formula: (EAD - Gross Post-Default Recoveries) / EAD
# If EAD is £10,000 and collection agencies claw back £2,000, the loss is 80% (0.80)
processed_df.loc[default_mask, 'LGD'] = (
    (processed_df.loc[default_mask, 'EAD'] - processed_df.loc[default_mask, 'recoveries']) 
    / processed_df.loc[default_mask, 'EAD']
)

# Constrain LGD bounds cleanly between 0.0 (Full Recovery) and 1.0 (Total Loss)
processed_df.loc[default_mask, 'LGD'] = processed_df.loc[default_mask, 'LGD'].clip(0.0, 1.0)

# For healthy loans (TARGET == 0), set LGD explicitly to 0.0 to avoid parsing issues
processed_df.loc[processed_df['TARGET'] == 0, 'LGD'] = 0.0

# ==========================================
# 📊 FRAMEWORK AUDIT SUMMARY REPORT
# ==========================================
print("🏦 ==========================================================")
print("            TARGET ENGINEERING SCHEME INITIALIZATION          ")
print("==============================================================")
print(f"✅ Total Cleaned Rows Extracted       : {processed_df.shape[0]}")
print(f"💳 Total Portfolio Outright Defaults  : {processed_df['TARGET'].sum()}")
print(f"📊 Dataset Default Density (PD Base)  : {processed_df['TARGET'].mean() * 100:.2f}%")
print("-" * 62)

default_only_df = processed_df[processed_df['TARGET'] == 1]
print(f"📉 Mean EAD Exposure Value on Default : £{default_only_df['EAD'].mean():,.2f}")
print(f"🔥 Mean LGD Loss Severity Percentage : {default_only_df['LGD'].mean() * 100:.2f}% loss")
print("==============================================================\n")
```

```python
import numpy as np
import pandas as pd
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# ==============================================================================
# 1. SETUP DATA AND CONFIGURATIONS FOR EAD
# ==============================================================================
# Filter out healthy loans: EAD operates strictly on historic default events
ead_dataset = processed_df[processed_df['TARGET'] == 1].copy()

# Ensure the target contains no missing values
ead_dataset = ead_dataset.dropna(subset=['EAD']).copy()

final_ead_clean_exclusions = [
    'TARGET', 'LGD', 'EAD',                       # System targets
    'issue_d', 'earliest_cr_line',               # Raw unparsed datetime strings
    'sec_app_earliest_cr_line',                  # Secondary datetime string
    'term', 'grade', 'sub_grade',                # Shortcuts/Dilutions
    
    # --- ALL POST-ORIGINATION HARDSHIP LEAKS (PURGED) ---
    'hardship_flag', 'hardship_type', 'hardship_reason', 'hardship_status', 
    'deferral_term', 'hardship_amount', 'hardship_start_date', 'hardship_end_date', 
    'payment_plan_start_date', 'hardship_length', 'hardship_dpd', 'hardship_loan_status', 
    'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount', 
    'hardship_last_payment_amount',
    
    # --- ALL POST-ORIGINATION SETTLEMENT LEAKS (PURGED) ---
    'debt_settlement_flag', 'debt_settlement_flag_date', 'settlement_status', 
    'settlement_date', 'settlement_amount', 'settlement_percentage', 'settlement_term',
    
    # --- ACCOUNT LIFECYCLE REPAYMENT LEAKS ---
    'total_rec_prncp', 'total_pymnt', 'total_pymnt_inv', 
    'total_rec_int', 'total_rec_late_fee', 'recoveries', 
    'collection_recovery_fee', 'last_pymnt_d', 'last_pymnt_amnt', 
    'next_pymnt_d', 'last_credit_pull_d', 'out_prncp', 'out_prncp_inv'
]

# Re-filter your training columns strictly using this true day-one baseline
clean_underwriting_cols = [col for col in ead_dataset.columns if col not in final_ead_clean_exclusions]

X = ead_dataset[clean_underwriting_cols]
y = ead_dataset['EAD'].astype(float)

# Automatically register remaining day-one categoricals
cat_cols = X.select_dtypes(include=['object', 'category', 'string']).columns.tolist()
for col in cat_cols:
    X[col] = X[col].astype(str).fillna('missing')

print(f"✅ Definitively Cleansed of Look-Ahead Bias.")
print(f"📊 Pure Underwriting Predictors Remaining: {X.shape[1]}")


print(f"✂️ Leakage features purged successfully.")
print(f"📊 Training Matrix Size: {X.shape} Defaulted Loans | {X.shape} Predictors.")
print(f"🗂️ Remaining Categorical Features: {len(cat_cols)}")

# ==============================================================================
# 2. DEFINING SOTA EAD HYPERPARAMETERS
# ==============================================================================
sota_ead_params = {
    'iterations': 1200,                  
    'learning_rate': 0.02,               
    'depth': 5,                          
    'l2_leaf_reg': 20,                   # Mitigates feature dilution over your 130+ fields
    'loss_function': 'RMSE',             # Optimizes continuous pound-value variation
    'eval_metric': 'R2',                 # Tracks explained exposure variance
    'random_seed': 42,
    'early_stopping_rounds': 50,
    'verbose': 0                         
}

# ==============================================================================
# 3. K-FOLD CROSS-VALIDATION LOOP
# ==============================================================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold_metrics = []

# Save references from the final fold loop for feature importance and SHAP analysis
final_trained_model = None
last_val_idx = None

print("\n🚀 Commencing 5-Fold EAD Regression Run...")

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y), 1):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    train_pool = Pool(X_train, y_train, cat_features=cat_cols)
    val_pool = Pool(X_val, y_val, cat_features=cat_cols)
    
    # Initialize and train the Regressor
    model = CatBoostRegressor(**sota_ead_params)
    model.fit(train_pool, eval_set=val_pool, use_best_model=True)
    
    # Generate predictions and clip floor at 0 to avoid impossible negative exposures
    val_preds = np.clip(model.predict(val_pool), 0.0, None)
    
    # Hold references securely
    final_trained_model = model
    last_val_idx = val_idx
    
    r2 = r2_score(y_val, val_preds)
    rmse = np.sqrt(mean_squared_error(y_val, val_preds))
    mae = mean_absolute_error(y_val, val_preds)
    
    fold_metrics.append({'R2': r2, 'RMSE': rmse, 'MAE': mae})
    print(f"  ↳ Fold {fold} Metrics -> R²: {r2:.4f} | RMSE: £{rmse:,.2f} | MAE: £{mae:,.2f}")

# ==============================================================================
# 4. FINAL REGRESSION PERFORMANCE SUMMARY
# ==============================================================================
print("\n================================================================================")
print("FINAL EAD MODEL SUMMARY")
print("================================================================================")
print(f"MEAN R² SCORE (Explained Variance) : {np.mean([f['R2'] for f in fold_metrics]):.4f}")
print(f"MEAN RMSE (Avg Value Deviation)    : £{np.mean([f['RMSE'] for f in fold_metrics]):,.2f}")
print(f"MEAN MAE (Mean Absolute Error)     : £{np.mean([f['MAE'] for f in fold_metrics]):,.2f}")
print(f"R² ST. DEV.                        : {np.std([f['R2'] for f in fold_metrics]):.4f}")
print("================================================================================\n")

# ==============================================================================
# 5. SAFE GLOBAL FEATURE IMPORTANCE EXTRACTION
# ==============================================================================
# Extract values and properties directly from the final_trained_model object to prevent mismatch errors
global_clean_ead_importances = final_trained_model.get_feature_importance()
model_feature_names = final_trained_model.feature_names_

ead_clean_importance_df = pd.DataFrame({
    'Feature_Variable_Name': model_feature_names,
    'Global_Weight_Percentage': global_clean_ead_importances
})

ead_clean_importance_df = ead_clean_importance_df.sort_values(
    by='Global_Weight_Percentage', 
    ascending=False
).reset_index(drop=True)

print("🏆 ==========================================================")
print("          CLEAN GLOBAL EAD FEATURE IMPORTANCE RANKING       ")
print("==============================================================")
print(f"{'Rank':<8}{'Feature Name':<35}{'Weight (%)'}")
print("-" * 62)

for idx, row in ead_clean_importance_df.head(15).iterrows():
    print(f"#{idx+1:<7}{row['Feature_Variable_Name']:<35}{row['Global_Weight_Percentage']:.4f}%")
print("==============================================================\n")
```

### 📊 EAD Feature Dictionary & Mathematical Specifications

The table below summarizes a highly predictive selection of the **132 pure behavioral features** utilized in the leak-free, state-of-the-art CatBoost Exposure at Default (EAD) model. Because this EAD regression pipeline operates exclusively on accounts that have already failed (`TARGET == 1`), these features are structured to calculate the speed of principal amortization and early-lifecycle default triggers using **only day-one origination metadata**.


| Feature Name | Core Business Meaning | Mathematical Formula / Algorithmic Representation |
| :--- | :--- | :--- |
| **`loan_amnt`** | The total nominal value of the loan requested by the applicant. | Input Column: $A_{loan} \in \mathbb{R}^{+}$ |
| **`funded_amnt`** | The total amount of principal committed to the loan by the bank. | Input Column: $A_{funded} \in \mathbb{R}^{+}$ |
| **`funded_amnt_inv`** | The portion of the loan principal funded explicitly by individual platform investors. | Input Column: $A_{inv} \in \mathbb{R}^{+}$ |
| **`installment`** | The fixed contractual monthly payment owed by the borrower. | $I = A_{loan} \times \frac{r(1+r)^N}{(1+r)^N - 1}$ <br> *(Where $r = \text{monthly interest rate}$, $N = \text{term months}$)* |
| **`int_rate`** | The annualized interest rate charged on the credit contract. | Input Column: $R_{annual} \in \mathbb{R}^{+}$ |
| **`annual_inc`** | The borrower's self-reported or verified gross annual income. | Input Column: $I_{annual} \in \mathbb{R}^{+}$ |
| **`monthly_income`** | The normalized monthly earning potential of the applicant. | $I_{monthly} = \frac{\text{annual\_inc}}{12}$ |
| **`emp_length`** | Employment longevity in years; proxies employment and income stability. | Categorical Mapping: $E \in \{\text{< 1 year}, \text{1 year}, \dots, \text{10+ years}\}$ |
| **`purpose`** | The stated intent of borrowing (e.g., debt consolidation, credit card payoff). | Categorical Mapping: $P \in \{\text{debt\_consolidation}, \text{credit\_card}, \dots\}$ |
| **`last_fico_range_low`** | The lowest boundary of the borrower's FICO score pulled during the default window. | Input Column: $\text{FICO}_{low} \in [300, 850]$ |
| **`last_fico_range_high`** | The upper boundary of the borrower's FICO score pulled during the default window. | Input Column: $\text{FICO}_{high} \in [300, 850]$ |
| **`num_rev_tl_bal_gt_0`** | Total number of revolving accounts currently carrying a non-zero balance. | Input Column: $N_{active\_rev} \in \mathbb{Z}^{+}$ |
| **`loan_to_income`** | Leverage ratio; sizes the total principal obligation against annual salary. | $\text{LTI} = \frac{\text{loan\_amnt}}{\text{annual\_inc}}$ |
| **`installment_to_income`** | The immediate contractual cash drain on the borrower's monthly paycheck. | $\text{ITI} = \frac{\text{installment}}{I_{monthly}}$ |
| **`term_loan_burden`** | Fixed loan concentration adjusted across remaining repayment timelines. | $\text{TLB} = \frac{\text{installment} \times \text{term\_months}}{\text{annual\_inc}}$ |
| **`active_trade_ratio`** | Tracks how many available revolving lines are actively carrying a balance. | $\text{ATR} = \frac{\text{num\_rev\_tl\_bal\_gt\_0}}{\text{num\_op\_rev\_tl} + 1}$ |
| **`open_account_ratio`** | Measures the proportion of accounts currently active out of total historical accounts. | $\text{OAR} = \frac{\text{open\_acc}}{\text{total\_acc} + 1}$ |
| **`util_per_account`** | Identifies if balances are evenly spread or dangerously maxed on single lines. | $\text{UPA} = \frac{\text{revol\_util}}{\text{open\_acc} + 1}$ |
| **`credit_hunger`** | Captures urgent credit-seeking velocity over the last 6 months. | $\text{CH} = \frac{\text{inq\_last\_6mths}}{\text{open\_acc} + 1}$ |
| **`payment_stress_amplification`** | Core dynamic compound multiplier tracking extreme default risk profiles. | $\text{PSA} = \text{installment\_to\_income} \times \text{utilization\_dti\_stress}$ |
| **`debt_saturation`** | Structural leverage stress tracking total credit usage relative to capacity. | $\text{DS} = \text{dti} \times \text{revol\_util}$ |
| **`long_term_stress`** | Gauges exposure to systemic long-term repayment friction. | $\text{LTS} = \text{dti} \times \text{mo\_sin\_old\_il\_acct}$ |
| **`loan_per_fico`** | Sizes requested asset concentration relative to historical credit tiering. | $\text{LPF} = \frac{\text{loan\_amnt}}{\text{fico\_average}}$ |

```python
import numpy as np
import pandas as pd
from catboost import Pool

def generate_master_expected_loss_dashboard(df, pd_model, lgd_model, ead_model):
    """
    Unifies PD, LGD, and EAD models to calculate the legal and financial
    Expected Loss (EL) in pounds for an institutional credit portfolio.
    """
    # Create a deep working copy for the financial pricing engine
    master_df = df.copy()
    
    # 1. Extract exact feature lists historically required by each individual model
    pd_features  = pd_model.feature_names_
    lgd_features = lgd_model.feature_names_
    ead_features = ead_model.feature_names_
    
    # 2. Automatically discover and cast categorical fields for safety across all frames
    def process_model_cats(dataframe, feature_list):
        sub_df = dataframe[feature_list].copy()
        cats = sub_df.select_dtypes(include=['object', 'category', 'string']).columns.tolist()
        for col in cats:
            sub_df[col] = sub_df[col].astype(str).fillna('missing')
        return sub_df, cats

    X_pd,  cat_pd  = process_model_cats(master_df, pd_features)
    X_lgd, cat_lgd = process_model_cats(master_df, lgd_features)
    X_ead, cat_ead = process_model_cats(master_df, ead_features)
    
    print("⏳ Processing pipeline predictions across multi-model architecture...")
    
    # 3. Step 1: Run Probability of Default (PD) Classifier
    pd_pool = Pool(X_pd, cat_features=cat_pd)
    master_df['Predicted_PD'] = pd_model.predict_proba(pd_pool)[:, 1]
    
    # 4. Step 2: Run Loss Given Default (LGD) Regressor
    lgd_pool = Pool(X_lgd, cat_features=cat_lgd)
    master_df['Predicted_LGD'] = np.clip(lgd_model.predict(lgd_pool), 0.0, 1.0) # Clip bounds securely [0%, 100%]
    
    # 5. Step 3: Run Exposure at Default (EAD) Regressor
    ead_pool = Pool(X_ead, cat_features=cat_ead)
    master_df['Predicted_EAD'] = np.clip(ead_model.predict(ead_pool), 0.0, None) # Floor at £0
    
    # 6. Step 4: Basel III Expected Loss Calculation Matrix
    # Formula: EL = PD * LGD * EAD
    master_df['Expected_Loss_£'] = (
        master_df['Predicted_PD'] * 
        master_df['Predicted_LGD'] * 
        master_df['Predicted_EAD']
    )
    
    # Calculate the normalized risk premium rate (EL as a % of starting funded capital)
    master_df['Expected_Loss_Rate_%'] = (master_df['Expected_Loss_£'] / master_df['funded_amnt']) * 100
    
    # ==============================================================================
    # 📊 REGULATORY CAPITAL AND PORTFOLIO PRICING REPORT
    # ==============================================================================
    total_funded = master_df['funded_amnt'].sum()
    total_ead    = master_df['Predicted_EAD'].sum()
    total_el     = master_df['Expected_Loss_£'].sum()
    global_el_rate = (total_el / total_funded) * 100
    
    print("\n🏦 ==========================================================")
    print("            BASEL III PORTFOLIO EXPECTED LOSS REPORT          ")
    print("==============================================================")
    print(f"💰 Total Underwritten Book Volume      : £{total_funded:,.2f}")
    print(f"📉 Total Forecasted Exposure (EAD Pool): £{total_ead:,.2f}")
    print(f"🚨 Total Required Capital Provision(EL): £{total_el:,.2f}")
    print(f"🎚️ Global Weighted Expected Loss Rate  : {global_el_rate:.2f}%")
    print("==============================================================")
    
    # Print the top 3 highest absolute financial risk pockets
    print("\n🔥 CRITICAL RISK CONCENTRATIONS (TOP 3 HIGHEST EXPECTED LOSS ASSETS):")
    print("-" * 62)
    top_risk = master_df.sort_values(by='Expected_Loss_£', ascending=False).head(3)
    for idx, row in top_risk.iterrows():
        print(f"Loan Index: {idx:<5} | Sub-Grade: {row['sub_grade']:<3} | Funded: £{row['funded_amnt']:<7,.0f}")
        print(f"  ↳ Metrics -> PD: {row['Predicted_PD']*100:.1f}% | LGD: {row['Predicted_LGD']*100:.1f}% | EAD: £{row['Predicted_EAD']:,.2f}")
        print(f"  ↳ IMPACT  -> Target Provisioning Capital Required: £{row['Expected_Loss_£']:,.2f}\n")
    print("==============================================================\n")
    
    return master_df

# 🚀 Active Execution Command Line:
# 🚀 Active Execution Command Line (Updated with your active DataFrame names):
final_audit_dashboard = generate_master_expected_loss_dashboard(
    df=processed_df,       # Uses the unified dataset holding your fresh targets
    pd_model=model,     # Your trained CatBoost PD Classifier
    lgd_model=model_lgd,   # Your trained CatBoost LGD Regressor 
    ead_model=model        # Your trained CatBoost EAD Regressor (current model)
)

```
